# Estructuras de datos
Las estructuras de datos son formas de organizar informacion para trabajar con ellas de manera eficiente

- Listas -> ordenadas, modificadas, permiten duplicados
- Tuplas -> ordenadas, inmutables, permiten duplicados
- Sets -> no ordenados, sin duplicados
- Diccionarios -> Pares clave:valor

Nota:

- Pilas (FILO -> First in, Last out)
- Colas (FILO -> First in, Last out)
- Listas enlazadas (o doblemente enlazadas)

## Listas
Las listas son colecciones ordenadas que podemos modificar y almacenan elementos de distintos tipos de datos

```python
lista_de_compras = ["manzanas","pan","leche","huevos","mantequilla","queso","jamon"]
print(lista_de_compras)
```

```python
print(lista_de_compras[6])
```

```python
# Agregar un elemento a la lista .append()
lista_de_compras.append("sopa")
lista_de_compras
```

```python
# Agegar un elemento en una posicion especifica
lista_de_compras.insert(1,"platano") # Esto lo duplica en la lista si lo ejecutas mas de dos veces
lista_de_compras
```

```python
print(lista_de_compras)
```

```python
# Para eliminar un elemento de una lista existe el remove y el pop
lista_de_compras.remove("platano")
print(lista_de_compras)
```

```python
lista_de_compras.pop() # Sin nada dentro del parentesis borra el ultimo elemento
lista_de_compras.pop(7) # o puedes agarrar la posicion que quieras (pero iniciando desde 0)
# Nota: te retona el numero eliminado
```

```python
# Podemos recorrer el areglo por medio de un for
for elementos in lista_de_compras:
  print(elementos)
```

```python
# len() saca el tama√±o de la lista
print(len(lista_de_compras))
for i in range(len(lista_de_compras)):
  print(f"{i + 1} - {lista_de_compras[i]}")
```

```python
promedios = [55.6,97.5,23.6,97.5,72.4,100]
print(promedios)
promedios.sort() # .sort() sirve para ordenar la lista de manera ascendente
print(promedios)
# ordenar una lista de manera descendente
promedios.sort(reverse = True)
print(promedios)
```

```python
# Podemos ingresar al ultimo elemento de una lista
print(promedios[-1])
```

```python
# Encontrar el mayor y el menor de una lista
print(max(promedios))
print(min(promedios))
```

```python
# Declaracion y uso de matrices
calificaciones = [[100,88,93],[99,56,72]]
i = 1
for cali_i in calificaciones:
  print(f"\nAlumno {i}")
  j = 1
  for cali_j in cali_i:
    print(f"Cali {j}:{cali_j}")
    j += 1
  j = 0
  i += 1
```

## SETS
Los sets son colecciones SIN orden y que NO permiten elementos duplicados

```python
nombre = {"Angel","Eliud","Diana","Juan","Juan"}
print(nombre)
print(type(nombre))
```

```python
distintos_datos = {"Angel", 1,"Diana","Juan",4.8}
distintos_datos
```

```python
# Add
nombre.add("Leonardo")
nombre
```

```python
print(nombre[2])
```

```python
nombre[2] = "Axel"
```

```python
# Union
matutino = {20322,22183,21547,256548}
vespertino = {20322,11111,22222,33333}
# Union pero con .union
total_matriculas = matutino.union(vespertino)
print(total_matriculas)
# Union pero con |
total_matriculas = matutino | (vespertino)
print(total_matriculas)
```

```python
# Interseccion
ca = {1234,121,646,12414,73434,124124,575234}
al = {1234,11,22,33,444}
# Interseccion con .intersection()
ca_al = ca.intersection(al)
print(ca_al)
# Interseccion usando &
ca_al = ca & al
print(ca_al)
```

```python
# Diferencia
# Diferencia usando .difference
solo_calculo = ca.difference(al)
print(solo_calculo)
# Diferencia usando -
solo_calculo = ca - al
print(solo_calculo)
```

```python
nombre = {"Angel","Eliud","Diana","Juan","Juan"}
# remove, pop
nombre.remove("Angel") # Puedes eliminar el que tu quieres
print(nombre)
nombre.pop() # Elimina un elemento de manera aleatoria
print(nombre)
```

```python
# Elimina todo el SET
nombre.clear()
nombre
```

## Diccionarios
Los diccionarios son colecciones no ordenadas de pares clave:valor

```python
alumno = {
    #clave      valor
    "nombre": "Jose Carlos",
    "edad": 14,
    "carrera": "Bachillerato",
    "semestre": 1,
    "promedio": 88,
    "calificaciones": [100,94,94,97,100]
}
```

```python
print("El alumno", alumno["nombre"], "esta estudiando", alumno["carrera"])
```

```python
# Agregar al diccionario
alumno["area preferida"] = "Ciencia de Datos"
alumno
```

```python
for valor in alumno:
  print(alumno[valor])
```

```python
for clave, valor in alumno.items() # Devuelve una vista de todos los pares clave:valor del diccionario
print(clave,":",valor)
```

```python
# El metodo .clear() tambien usar aqui
# Temenos el .get que nos ayuda a obtener el valor (value) de cierta clave (key)
print(alumno.get("nombre"))
```

```python
alumno_1 = alumno.items()
print(alumno_1)
print(type(alumno_1))
alumno_lista = list(alumno_1)
print(alumno_lista)
print(type(alumno_lista))
```

```python
# El metodo .values() sirve para obtener los valores del diccionario, tambien se puede usar la interaccion
alumno_v = alumno.values()
print(type(alumno_v))
alumno_v
```

```python
if "carrera" in alumno:
  print("Existe carrera")
else:
  print("No existe carrera")

if("dinero") in alumno:
  print("Alumno tiene dinero")
else:
  print("Alumno no tiene dinero")
```

```python
#.copy() puedes copiar todo el diccionario en una nueva variable
# Update -> Sirve para cambiar los values (valores) (.update())
alumno.update({"semestre":2})
alumno
```
