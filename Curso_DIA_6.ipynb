- Recibe como primer parametro a self

```python
# Calculadora:
class Calculadora:
  def __init__(self,num1,num2):
    self.num1 = num1
    self.num2 = num2
  def sumar(self):
   print(self.num1 + self.num2)
  def resta(self):
   print(self.num1 - self.num2)
  def multiplicar(self):
    print(self.num1 * self.num2)
  def dividir(self):
   if self.num2 == 0:
      print("Numero no divisible")
   else:
     print(self.num1/self.num2)
Calculadora = Calculadora(7,0)
Calculadora.dividir()
Calculadora.sumar()
Calculadora.multiplicar()
```

## Encapsulamiento y propiedades

```python
class Cuenta:
  def __init__(self,saldo):
    self.__saldo = saldo # Privada
# Sirve para protejer los datos internos de un objeto y controlar como se leen o modifican
cuenta = Cuenta(1000)
cuenta.Cuenta = 2000 # Si puedes editarlo
# print(cuenta.__saldo) # Error no se puede acceder directamente
```

```python
# Getters y Setters
class Cuenta:
  def __init__(self,saldo):
    self.__saldo = saldo
  @property
  def saldo(self): # Getter
  # Permite leer el valor de _saldo usando c.saldo
    return self.__saldo
  @saldo.setter # Setter
  def saldo(self,valor):
    # se ejecuta cuendo hacemmos c.saldo = valor
    self.saldo = valor
c = Cuenta(300)
print(c.saldo)
```

## Herencia y abstraccion

```python
# Herencia
class vehiculo: # Clase base (Padre)
  def __init__(self, marca, modelo):
    self.marca = marca
    self.modelo = modelo
  def descripcion(self):
    return f"{self.marca}:{self.modelo}"
  def mover(self):
    return "El vehiculo se esta moviendo..."
# -------------------------------------------------------
class Coche(vehiculo): # Subclase 1 (Clase hija)
  def maver(self):
    return "El coche esta conduciendo"
# -------------------------------------------------------
class Moto(vehiculo):
  def mover(self):
    return "La moto avanza sobre dos ruedas"
coche2 = vehiculo("BYD", "Dolphin")
coche = Coche("Toyota", "Corolla")
moto = Moto("Yamaha", "R6")
print(coche2.mover())
print(coche.mover()) #aun que no volvimos a declarar el metodo dentro de descripcion, podemos seguir usandolo ya que lo heredo del padre
print(moto.mover())
```

```python
class Persona:
  def __init__(self,nombre,edad):
    self.nombre = nombre
    self.edad = edad
  def presentarse(self):
    print(f"Hola soy {self.nombre} y tengo {self.edad} a単os")
class Estudiante(Persona):
  def __init__(self, nombre, edad, carrera):
    super().__init__(nombre, edad)
    self.carrera = carrera
  def presentarse(self):
    super().presentarse()
    print(f"Estudio la carrera de {self.carrera}")
  # Reutiliza el constructor de la clase padre (lo cual evita duplicar el codigo)
  # Ejecuta el comportamiento base y luego lo extiende
est = Estudiante("Ana", 21, "Programacion")
est.presentarse()
```

## Composicion HAS-A
Por que se llama composicion?
Por que estas componiendo (armando) un objeto grande usando otros objetos. Es un principio de POO donde:
- Una clase NO hereda de otra, sino que contiene un objeto de esta otra clase

```python
class Banco:
  def __init__(self,dinero):
    self.dinero = dinero
  def ver_total(self):
    print("Total en la cuenta: ${self.dinero}")
class Persona:
  def __init__(self, cuenta: Banco): #Indica que este parametro deberia ser una instancia de la clse Banco
    self.cuenta = cuenta # Persona TIENE un banco, no hereda de el
```

## Composision vs Herencia
Esta se va a responder con pura teoria:
No significa que alguna sea mejor que otra, simplemente que aveces una te puede convenir mas que otra:

Herencia: un gato ES un animal (no "Un gato TIENE un animal")

Composicion: Una persona TIENE una cuenta (no "Una persona ES un banco")
