## Introduccion a la Regresion Lineal

```python
from sklearn.linear_model import LinearRegression
```

### Â¿Que es una regresion lineal?
Una regresion intenta predecir un valor numerico a partir de otros datos

Ejemplo:
- Predecir ingresos segun edad
- Predecir precio segun tamaÃ±o

Aqui usaremos regresion lineal: el tipo mas simple

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x = df_clean["edad"], y = df_clean["ingresos"])
plt.xlabel("Edad")
plt.ylabel("Ingresos")
plt.title("Relacion entre Edad y Ingresos")
plt.show()
```

```python
# MATRIZ DE CORRELACION
#df_clean.corr()
df_clean[["edad", "ingresos"]].corr()

# Niveles de correlacion
# 0.0 -> nada
# 0.3 -> debil
# 0.5 - 0.7 -> moderada
# > 0.8 -> fuerte
```

```python
print(df_clean)
```

```python
x = df_clean[["edad"]]
y = df_clean["ingresos"]
modelo = LinearRegression()
modelo.fit(x, y)
```

Tiene relacion con la formula de la pendiente

Â¿Que hace fit?
- Busca los valores de m, y b que mejor se ajustan a tus datos. Minimiza el error entre los ingresos reales y los ingresos predichos

```python
edad_nueva = [[30]]
prediccion = modelo.predict(edad_nueva)
prediccion
```

```python
# Podemos evaluar el modelo
from sklearn.metrics import r2_score, mean_squared_error
y_pred = modelo.predict(x) # Son los ingresos que el modelo cree segun la edad
#ingresos = Î²0 + Î²1 Â· edad
#Î²0 â intercepto
#Î²1 â coeficiente de edad

r2 = r2_score(y, y_pred)
# 0.3 - 0.5 aceptable, pero siempre es mejor tener un modelo 0.6 - 0.8 o muy bueno > 0.8
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)
print(r2)
# Regresion Logistica se suele usar -> Acccuracy, Precision, F1 score, Recall, Roc Auc
```
