## Funciones que llaman a otras funciones

```python
def saludar(nombre):
  return f"Hola, {nombre}!"

def despedirse(nombre):
  return f"Adios, {nombre}!"

def mensaje_completo(nombre):
  saludo = saludar(nombre)
  despedida = despedirse(nombre)
  return saludo + " " + despedida
print(mensaje_completo("Ana"))
```

```python
# Ahora supongamos que tenemos una heladeria
def pedir_sabor():
  sabores = ["vainilla", "chocolate", "fresa"]
  print("Sabores disponible", ",", sabores)
  # lower() hace minusculas las palabras
  # strip() elimina los espacios tanto antes como despues
  sabor = input("Elige un sabor: ").strip().lower()
# si sabor NO esta en sabores
  while sabor not in sabores:
    sabor = input("Elige un sabor (nuevamente): ").strip().lower()
  return sabor

def pedir_cantidad():
  cantidad = int(input("Cuantas bolas de nieve?: "))
  while cantidad <= 0:
     cantidad = int(input("Cuantas bolas de nieve? (ingresa un valor valido): "))
  return cantidad
  # Funcion Madre (llama a otras funciones)
def pedir_helado():
  sabor = pedir_sabor()
  cant = pedir_cantidad()
  print(f"Pedido: {cant} bola(s) de {sabor}")
  return cant, sabor

total_pedidos = int(input("Cuantos clientes atenderas?: "))
for i in range(total_pedidos):
  pedir_helado()
```

## Funciones Lambda
Es una funcion acortada, para escribir una funcion es de esta forma:

nombreFuncion = lamdba parametros: accion

La diferencia entre estas y las funciones normales es que suelen ser un poco mas limitadas

```python
sumar = lambda x,y : x+y
print(sumar(5,9))
```

```python
descuentoLambda = lambda desc, precio: ((precio * (1-desc)), precio * desc)
nuevoPrecio = descuentoLambda(0.4,500)
print("El precio a pagar es", nuevoPrecio[0], ", el descuento que se hizo fue de:", nuevoPrecio)
print(type(nuevoPrecio))
```

```python
# Calculadora -> suma, resta, producto, division y potencia
def suma(a,b):
  return a + b
def resta(a, b):
  return a - b
def producto(a,b):
  return a * b
def division(a,b):
  if b == 0:
    return "Error: no se puede dividir entre 0"
  return a / b
def potencia(a,b):
  return a ** b # Corrected: 'a' is the base, '**' is the operator, 'b' is the exponent

def menu():
  print("---------- CALCULADORA ---------")
  print("1. suma")
  print("2. Resta")
  print("3. Producto")
  print("4. division")
  print("5. Potencia")
  print("6. Salir")

while True:
  menu()
  try:
    op = int(input("Seleccione una opcion: "))
    if op < 1 or op > 6:
      print("Opcion no valida")
      continue # Restart the loop if option is invalid
    elif op == 6:
      print("SALIENDO...")
      break # Exit the loop if user chooses to exit
  except ValueError:
    print("Ingrese un numero invalido")
    continue # Restart the loop if non-integer is entered for option

  try:
    a = float(input("Ingrese el primer numero: ")) # Corrected typo from 'numero;' to 'numero:'
    b = float(input("Ingrese el segundo numero: "))
  except ValueError:
    print("Error: Ingrese un numero valido")
    continue # Restart the loop if non-numeric input for operands

  if op == 1:
    print("Resultado: ", suma(a,b))
  elif op == 2:
    print("Resultado: ", resta(a,b))
  elif op == 3:
    print("Resultado: ", producto(a,b))
  elif op == 4:
    print("Resultado: ", division(a,b)) # Corrected: pass a and b as separate arguments
  elif op == 5:
    print("Resultado: ", potencia(a,b))
  print("-" * 30) # Add a separator for better readability between operations
```

```python
# Calculadora -> suma, resta, producto, division y potencia
def suma(a, b):
    return a + b

def resta(a, b):
    return a - b

def producto(a, b):
    return a * b

def division(a, b):
    if b == 0:
        return "Error: no se puede dividir entre 0"
    return a / b

def potencia(a, b):
    return a ** b

def menu():
    print("-------- MI CALCULADORA --------")
    print("1. Suma")
    print("2. Resta")
    print("3. Producto")
    print("4. División")
    print("5. Potencia")
    print("6. Salir")

while True:
    menu()
    try:
        op = int(input("Seleccione una opción: "))
    except ValueError:
        print("Ingrese un número válido")
        continue

    if op == 6:
        print("SALIENDO...")
        break

    if op < 1 or op > 6:
        print("Opción inválida")
        continue

    try:
        a = float(input("Ingrese el primer número: "))
        b = float(input("Ingrese el segundo número: "))
    except ValueError:
        print("Error: Ingrese un número válido")
        continue

    if op == 1:
        print("Resultado:", suma(a, b))
    elif op == 2:
        print("Resultado:", resta(a, b))
    elif op == 3:
        print("Resultado:", producto(a, b))
    elif op == 4:
        print("Resultado:", division(a, b))
    elif op == 5:
        print("Resultado:", potencia(a, b))
```
